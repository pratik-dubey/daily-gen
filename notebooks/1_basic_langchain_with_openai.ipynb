#### Prompt Templates - Dynamic Prompts 

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

out = StrOutputParser()

llm = ChatGoogleGenerativeAI(model='gemini-2.5-flash-lite') 

prompt = ChatPromptTemplate.from_messages([
    {"role": "system", "content": "You are a translator who translates from input to {language}"},
    {"role":"user", "content": "{query}"}
])

# final_prompt = prompt.format_messages(language = "English", query = "i am 22 years old")
# res = llm.invoke(final_prompt)

# print(res.content)
```

```python
### My Self Chain Node

def transform_case(result:str):
    return result.upper()
```

```python
chains = prompt | llm | out | transform_case
print(chains.invoke({"language": "English", "query": "i love Langchain"}))
```
