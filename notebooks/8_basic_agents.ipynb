```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
# tool in langchain --> simple functions that ae defined by @tool decorator and inside the function we write the doc_string in """""" to give desc about that tool(function) 

@tool
def add_numbers(a:int, b:int):
    """
      it will add two numbers 

      Args:
         a: number one
         b: number two
    """

    return a + b
```

```python
@tool
def multiply_numbers(a:int, b:int):
    """
      it will multiply two numbers 

      Args:
         a: number one
         b: number two
    """

    return a * b
```

```python
add_numbers.invoke({"a":12, "b":3})
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
agent = create_agent(
    model=llm,
    tools=[add_numbers, multiply_numbers],
    system_prompt="Ypu are math teacher, and always use tool for calculation"
)
```

```python
res = agent.invoke({"messages":[{"role":"user", "content":"What is10 + 4*7"}]})
```
