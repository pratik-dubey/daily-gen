```python
from langchain_community.utilities import GoogleSerperAPIWrapper
from langchain_groq import ChatGroq
from langchain.agents import create_agent
# memory from langgraph
from langgraph.checkpoint.memory import MemorySaver
```

```python
llm = ChatGroq(model="openai/gpt-oss-20b")
search = GoogleSerperAPIWrapper()
```

```python
search.run("how many people were laid off by oracle recently ?")
```

```python
agent = create_agent(
    model=llm,
    tools=[search.run],
    system_prompt="You are a agent and can search for any question on google .",
    checkpointer=MemorySaver()
)
```

```python
question = "who is pm of india ?"

# this thread_id makes memory and if we keep asking questions without changing the thread id then llm will have the context of previous conversations
res = agent.invoke({"messages":{"role":"user", "content":question}},
       {"configurable":{"thread_id":"pk"}},            
)
```
