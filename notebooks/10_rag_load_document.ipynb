### Text Document Loader

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("../data/speech.txt", encoding="utf-8")
texts = loader.load()

# langchain gives list of document when we load any document and its length is equal to the number of pages in the document as it stores each page as a separate element of document

len(texts)
doc = texts[0]
# print(doc.page_content)
```
