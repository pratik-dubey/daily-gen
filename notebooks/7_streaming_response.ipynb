```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

### Streaming LLM response in real time

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", streaming=True)
```

```python
question = "who is gen ai ?"
# this llm.stream will only give a generator object and upon which we'll have to iterate 
response = llm.stream(question)
```

```python
# this llm.stream will only give a generator object and upon which we'll have to iterate 
response = llm.stream(question)
for chunk in response:
    print(chunk.content, end = "")

# end parameter decide karta hai ki print ke baad kya add hoga.
# Default: end="\n" (new line)
# Tumne diya: end="" (kuch bhi nahi)
```

### Streaming Responses using Groq

```python
from langchain_groq import ChatGroq
llm = ChatGroq(model="openai/gpt-oss-20b", streaming=True)
```

```python
question = "who is pm of india ?"
```

```python
res = llm.stream(question)

for chunk in res:
    print(chunk.content, end="")
```
