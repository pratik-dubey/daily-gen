```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
text = "Hello, i am Pratik Kumar Dubey , my email is pk@gmail.com and my number is 0909909090"
res = llm.invoke(f"please give me my name number and email from this test: {text}")
```

```python
from pydantic import Field, BaseModel
# pydantic is used for type validation in python
class ResponseStructure(BaseModel):
    name:str = Field(description="Complete name")
    email:str = Field(description="Email")
    age:str = Field(description="Age")

structured_llm = llm.with_structured_output(ResponseStructure)
```

```python
res = structured_llm.invoke(f"please give me my name number and email from this test: {text}")
```
