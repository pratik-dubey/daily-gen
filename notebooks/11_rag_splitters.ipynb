```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
texts = "Deep Agents are implementations of LangChain agents. If you don’t need these capabilities or would like to customize your own for your agents and autonomous applications, start with LangChain."
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size = 50, chunk_overlap = 20)

# we can use splitter.split_documents() for loaded docs (text, csv, pdf , etc.) 
chunk = splitter.split_text(texts)
chunk
```
